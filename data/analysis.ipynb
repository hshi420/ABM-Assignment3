```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv('seg_model_single_run_data1.csv')
df2 = pd.read_csv('seg_model_single_run_data2.csv')
df3 = pd.read_csv('seg_model_single_run_data3.csv')
```

```python
def plot_pct(df):
    plt.plot(df['Avg pct similar neighbors'], label = f"type_intol:{df['Intolerance'][0]},wealth_intol:{df['wealth_intolerance'][0]}")
    plt.xlabel('steps')
    plt.ylabel('Avg pct similar neighbors')
```

```python
df1['Intolerance'][0]
```

```python
plt.figure(dpi = 120)
plot_pct(df2)
plot_pct(df1)
plot_pct(df3)
plt.legend()
plt.title('single runs')
```

```python
df_batch = pd.read_csv('batch_run.csv')
```

```python
df_batch
```

```python
steps = list(range(0,51))*270
```

```python
df_batch['steps'] = steps
```

```python
from statistics import mean
```

```python
similarities = list(df_batch.groupby(['Intolerance','wealth_intolerance','steps']).mean()['Avg pct similar neighbors'])
```

```python
x = list(range(0,51))
```

```python
plt.figure(dpi = 200)
plt.plot(x, similarities[51:102], label = 'type_intol: 0.25,wealth_intol:0.5')
plt.xlabel('steps')
plt.ylabel('Avg pct similar neighbors')
plt.legend()
```
